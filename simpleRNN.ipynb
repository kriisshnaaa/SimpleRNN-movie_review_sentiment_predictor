Deep Learning project using simple RNN

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
## load imdb dataset
max_features=10000 #vocabulary size
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features) ##imdb movie ratings are in this dataset in the from of encoded values x have the review and y have is it good or bad
 
```

```python
max_len=500
X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test=sequence.pad_sequences(X_test,maxlen=max_len)
print(X_train.shape,X_test.shape)
```

```python
#train simple RNN
model=Sequential()
model.add(Embedding(max_features,128,input_length=X_train.shape[1])) ## taking all the words and converting it to the dimention of 128 dimention
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##create an instance of earlystopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
#train model
history=model.fit(X_train,y_train,epochs=10,batch_size=32,validation_split=0.2,callbacks=[early_stopping])
```

```python
loss,accuracy=model.evaluate(X_test,y_test)
```

```python
model.save('simpleRNN.h5')
```
